# Feature selection 
Firstly we need to identify the most important features for employee turnover. We'll be the some of the classifiers in the next section not for training purposes, but just for understanding which are the most important features that influence employee turnover. Here are the techniques we'll be using for feature selection -

- Co-relation heatmap
- Random Forest Classifier 
- XGBoost Classifier 
- Decision Trees (they reveal much more than that top features)

The heat-map earlier showed us that the `average monthly hour` feature is highly co-related to `number of projects` (obviously) and the evaluation. Due to this high co-relation this feature can be dropped while training with the classifiers in the next section.

## Preparing the data for training

```python
from sklearn.metrics import fbeta_score
from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split
import numpy as np
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (12,6)

df = data.drop(['last_evaluation_group','satisfaction_level_group'],axis=1)

# Convert these variables into categorical variables.
df["department"] = df["department"].astype('category').cat.codes
df["salary"] = df["salary"].astype('category').cat.codes

# Display the first 10 records
display(df.head(n=10))

# Create train and test splits
target_name = 'left'
X = df.drop('left', axis=1)

# obtain the target value. Its the employee turnover.
y = df[target_name]
# Split the data into test and training set.
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15, random_state=123, stratify=y)
```

### Random Forest Classifier for feature selection 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score, learning_curve

RFC = RandomForestClassifier()

RFC.fit(X_train,y_train)

rfc_predictions_train = RFC.predict(X_train)
rfc_predictions_test = RFC.predict(X_test)

rfc_training_accuracy_score = accuracy_score(y_train, rfc_predictions_train)
rfc_test_accuracy_score = accuracy_score(y_test, rfc_predictions_test)
print("Random Forest accuracy on training set:",rfc_training_accuracy_score)
print("Random Forest accuracy on testing set:",rfc_test_accuracy_score)

rfc_training_fbeta_score = fbeta_score(y_train, rfc_predictions_train, beta=0.5, average='micro')
rfc_test_fbeta_score = fbeta_score(y_test, rfc_predictions_test, beta=0.5, average='micro')
print("Random Forest fbeta score on training set:",rfc_training_fbeta_score)
print("Random Forest fbeta score on testing set:",rfc_test_fbeta_score)
```

```python
indices = np.argsort(RFC.feature_importances_)[::-1]

train_features = X_train.columns

print(train_features)
print(indices)

fig = plt.figure(figsize=(7, 10))
g = sns.barplot(y=train_features[indices],x = RFC.feature_importances_[indices] , orient='h')
g.set_xlabel("Relative importance",fontsize=12)
g.set_ylabel("Features",fontsize=12)
g.tick_params(labelsize=9)
g.set_title("Random Forest classifier feature importance")

plt.show()
plt.gcf().clear()
```

## XGBoost classifier for feature selection

```python
!pip install xgboost
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(X_train, y_train)

xgb_predictions_train = model.predict(X_train)
xgb_predictions_test = model.predict(X_test)

xgb_training_accuracy_score = accuracy_score(y_train, xgb_predictions_train)
xgb_test_accuracy_score = accuracy_score(y_test, xgb_predictions_test)
print("XGBoost accuracy on training set:",xgb_training_accuracy_score)
print("XGBoost accuracy on testing set:",xgb_test_accuracy_score)

xgb_training_fbeta_score = fbeta_score(y_train, xgb_predictions_train, beta=0.5, average='micro')
xgb_test_fbeta_score = fbeta_score(y_test, xgb_predictions_test, beta=0.5, average='micro')
print("XGBoost fbeta score on training set:",xgb_training_fbeta_score)
print("XGBoost fbeta score on testing set:",xgb_test_fbeta_score)
# feature importance
print(model.feature_importances_)

indices = np.argsort(model.feature_importances_)[::-1]

print(indices)

train_features = X_train.columns

print(train_features)
print(indices)

fig = plt.figure(figsize=(7, 10))
g = sns.barplot(y=train_features[indices],x = model.feature_importances_[indices] , orient='h')
g.set_xlabel("Relative importance",fontsize=12)
g.set_ylabel("Features",fontsize=12)
g.tick_params(labelsize=9)
g.set_title("XGBoost classifier feature importance")

plt.show()
plt.gcf().clear()
```

## K-Means clustering
Let's use K-means clustering on the data of people who left the company. Since we saw a clear grouping in employees who left the company in the `satisfaction vs evaluation` scatter plot, let's use these features to separate out the clusters and perform further analysis on each one of them.

```python
from sklearn.cluster import KMeans

# Obtain the data for the employees who've left the company.
data_left = df[df["left"]==1]

data_stayed = df[df["left"]==0]

print(data_left.shape)

kmeans = KMeans(n_clusters=3,random_state=2)
kmeans.fit(data_left[["satisfaction_level","last_evaluation"]])

```

```python
kmeans_colors = ['red' if c == 0 else 'orange' if c == 2 else 'blue' for c in kmeans.labels_]

fig = plt.figure(figsize=(10, 7))
plt.scatter(x="satisfaction_level",y="last_evaluation", data=data_left,
            alpha=0.25,color = kmeans_colors)
plt.xlabel("Satisfaction level")
plt.ylabel("Last evaluation")

clusters_center_x = kmeans.cluster_centers_[:,0]
clusters_center_y = kmeans.cluster_centers_[:,1]

print('cluster center 1: ', (clusters_center_x[0], clusters_center_y[0]))
print('cluster center 2: ', (clusters_center_x[1], clusters_center_y[1]))
print('cluster center 3: ', (clusters_center_x[2], clusters_center_y[2]))
plt.scatter(x = clusters_center_x, y = clusters_center_y, color="black",marker="X",s=100)
plt.title("Clustering of the employed who left by Kmeans")

plt.show()
```

Again, here are the 3 groups from the clusters.
- **Performing Unhappy**: Successful but unhappy employees. This is the yellow group on the top left
- **Performing Happy**: Successful and happy employees. These constitute the red group on the top right 
- **Non-performing Unhappy**: Unsuccessful and unhappy employees. This is the blue group on the bottom

```python
performing_satisfied_left = data_left[kmeans.labels_ == 0]
nonperforming_unsatisfied_left = data_left[kmeans.labels_ == 1]
performing_unsatisfied_left = data_left[kmeans.labels_ == 2]

print("Employees left: ")
print("Performing Happy: ",performing_satisfied_left.shape)
print("Non-performing Unhappy: ",nonperforming_unsatisfied_left.shape)
print("Performing UnHappy: ",performing_unsatisfied_left.shape)
```

### A Radar Chart on average features of employees who stayed and employees who left, from each of the three clusters,  should help us find the most important factors for employee turnover. 

> Note to self: Couldn't find a library to draw the radar chart, so I just used an implementation that I found online.

```python
# Code to draw Radar chart.
def _scale_data(data, ranges):
    (x1, x2) = ranges[0]
    d = data[0]
    return [(d - y1) / (y2 - y1) * (x2 - x1) + x1 for d, (y1, y2) in zip(data, ranges)]

  
class RadarChart():
    def __init__(self, fig, variables, ranges, n_ordinate_levels = 6):
        angles = np.arange(0, 360, 360./len(variables))

        axes = [fig.add_axes([0.1,0.1,0.8,0.8],polar = True, 
        label = "axes{}".format(i)) for i in range(len(variables))]
        _, text = axes[0].set_thetagrids(angles, labels = variables)
        
        for txt, angle in zip(text, angles):
            txt.set_rotation(angle - 90)
        
        for ax in axes[1:]:
            ax.patch.set_visible(False)
            ax.xaxis.set_visible(False)
            ax.grid("off")
        
        for i, ax in enumerate(axes):
            grid = np.linspace(*ranges[i],num = n_ordinate_levels)
            grid_label = [""]+["{:.1f}".format(x) for x in grid[1:]]
            ax.set_rgrids(grid, labels = grid_label, angle = angles[i])
            ax.set_ylim(*ranges[i])
        
        self.angle = np.deg2rad(np.r_[angles, angles[0]])
        self.ranges = ranges
        self.ax = axes[0]

    def plot(self, data, *args, **kw):
        sdata = _scale_data(data, self.ranges)
        self.ax.plot(self.angle, np.r_[sdata, sdata[0]], *args, **kw)

    def fill(self, data, *args, **kw):
        sdata = _scale_data(data, self.ranges)
        self.ax.fill(self.angle, np.r_[sdata, sdata[0]], *args, **kw)

    def legend(self, *args, **kw):
        self.ax.legend(*args, **kw)

```

```python
df["salary"] = df["salary"].astype('category').cat.codes
data_stayed["salary"] = data_stayed["salary"].astype('category').cat.codes
performing_unsatisfied_left["salary"] = performing_unsatisfied_left["salary"].astype('category').cat.codes
performing_satisfied_left["salary"] = performing_satisfied_left["salary"].astype('category').cat.codes
nonperforming_unsatisfied_left["salary"] = nonperforming_unsatisfied_left["salary"].astype('category').cat.codes




attributes = ['satisfaction_level','last_evaluation','total_projects','average_monthly_hours','years_at_company','salary']
ranges = [[1e-2, df[attr].max()] for attr in attributes]


fig = plt.figure(figsize=(8, 8))



stayed_mean = data_stayed[attributes].mean().values.reshape(1,-1)
performing_unsatisfied_left_mean = performing_unsatisfied_left[attributes].mean().values.reshape(1,-1)
performing_satisfied_left_mean = performing_satisfied_left[attributes].mean().values.reshape(1,-1)
nonPerforming_unsatisfied_left_mean = nonperforming_unsatisfied_left[attributes].mean().values.reshape(1,-1)


radar_data = np.concatenate((np.array(stayed_mean), np.array(performing_satisfied_left_mean), 
                      np.array(performing_unsatisfied_left_mean), np.array(nonPerforming_unsatisfied_left_mean)),axis=0)


colors = ["green","red","blue","orange"]
left_types = ["Stayed","Performing-Satisfied-Left", "Performing-Unsatisfied-Left", "NonPerforming-Unsatisfied-Left"]

fig = plt.figure(figsize=(17, 17))
radar = RadarChart(fig, attributes, ranges)
for data, color, left_type in zip(radar_data, colors, left_types):
    radar.plot(np.array(data), color = color, label = left_type, linewidth=2.0)
    radar.fill(data, alpha = 0.1, color = color)
    radar.legend(loc = 1, fontsize = 'small')
plt.title('Stats of employees who stayed and left')

plt.show() 
```

# Radar Graph interpretation

> ## Insights:
- The radar graph shows that the mean salary is the same for all cluster of employees. This clearly tells that the well performing employees need a raise. 
- It's clear why a large section of well performing employees are unsatisfied and they leave the  company. The blue part in the radar graph indicates that these employees work way too many projects and thus spend far too many hours at office but still get paid the same like the others. Pretty soon they become unsatisfied. It's clear that they might under lot of stress at work and their load has to reduce to improve the chances of retaining them.
